```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def np_move_avg(a,n,mode="same"):
    return(np.convolve(a, np.ones((n,))/n, mode=mode))
```

```python
cost3 = np.load('./data/data_result/cost_{}.npy'.format(3))
cost10 = np.load('./data/data_result/cost_{}.npy'.format(10))
cost9 = np.load('./data/data_result/cost_{}.npy'.format(9))
plt.figure(figsize=(10,6))
mode = 'valid'
cost3 = cost3[::60]
cost10 = cost10[::60]
cost9 = cost9[::60]
Step = 10000
plt.plot(np.linspace(0,Step,len(cost3)),cost3,alpha=0.95,c='r',label='Beijing')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost3,10, mode))),np_move_avg(cost3,10, mode),c='r',label='Beijing')
plt.plot(np.linspace(0,Step,len(cost10)),cost10,alpha=0.95,c='g',label='Guangzhou')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost10,10, mode))),np_move_avg(cost10,10, mode),c='g',label='Guangzhou')
plt.plot(np.linspace(0,Step,len(cost9)),cost9,alpha=0.95,label='Manhattan')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost9,10, mode))),np_move_avg(cost9,10, mode),c='b',label='Manhattan')
plt.legend()
plt.xlabel('Step')
plt.ylabel('Loss')
# plt.fill_between(smooth_x, smooth_y,alpha=0.2,color='r')
plt.savefig('./figures/Loss.png',dpi=500,bbox_inches='tight')
plt.show()
```

```python
reward3 = np.load('./data/data_result/reward_{}.npy'.format(3))
reward10 = np.load('./data/data_result/reward_{}.npy'.format(10))
reward9 = np.load('./data/data_result/reward_{}.npy'.format(9))
print(len(reward3))
plt.figure(figsize=(10,6))
mode = 'valid'
reward3 = reward3[::3]
reward10 = reward10[::3]
reward9 = reward9[::3]
Episode = 600
plt.plot(np.linspace(0,Episode,len(reward3)),reward3,alpha=0.2,c='r')
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward3,10, mode))),np_move_avg(reward3,10, mode),c='r',label='Beijing')
plt.plot(np.linspace(0,Episode,len(reward10)),reward10,alpha=0.2,c='g')
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward10,10, mode))),np_move_avg(reward10,10, mode),c='g',label='Guangzhou')
plt.plot(np.linspace(0,Episode,len(reward9)),reward9,alpha=0.2)
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward9,10, mode))),np_move_avg(reward9,10, mode),c='b',label='Manhattan')
plt.legend()
plt.xlabel('Episode')
plt.ylabel('Reward')
# plt.fill_between(smooth_x, smooth_y,alpha=0.2,color='r')
plt.savefig('./figures/Reward.png',dpi=500,bbox_inches='tight')
plt.show()
```
