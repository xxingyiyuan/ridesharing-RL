```python
def np_move_avg(a,n,mode="same"):
    return(np.convolve(a, np.ones((n,))/n, mode=mode))
```

```python
cost3 = np.load('./data/data_result/cost_{}.npy'.format(3))
cost10 = np.load('./data/data_result/cost_{}.npy'.format(10))
cost9 = np.load('./data/data_result/cost_{}.npy'.format(9))
# Step = np.mean([len(cost3),len(cost10),len(cost9)])
Episode = 20000
plt.figure(figsize=(8, 6))
mode = 'valid'
step = 3
cost3 = cost3[::step]
cost10 = cost10[::step]
cost9 = cost9[::step]

# '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'
color1, color2, color3 ='#1f77b4', '#ff7f0e', '#2ca02c'
# color1, color2, color3 ='#9467bd', '#ff7f0e', '#1f77b4'
plt.plot(np.linspace(0,Episode,len(cost3)),cost3,alpha=1,c=color1,label='北京')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost3,10, mode))),np_move_avg(cost3,10, mode),c= color1)
plt.plot(np.linspace(0,Episode,len(cost10)),cost10,alpha=1,c=color2,label='广州')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost10,10, mode))),np_move_avg(cost10,10, mode),c=color2)
plt.plot(np.linspace(0,Episode,len(cost9)),cost9,alpha=1,c=color3,label='纽约')
# plt.plot(np.linspace(0,Step,len(np_move_avg(cost9,10, mode))),np_move_avg(cost9,10, mode),c=color3)
fontsize = 16
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.legend(fontsize=fontsize)
plt.xticks(fontsize=fontsize,fontfamily='Times New Roman')
plt.yticks(fontsize=fontsize,fontfamily='Times New Roman')
plt.xlabel('迭代次数',fontsize=fontsize)
plt.ylabel('损失',fontsize=fontsize)
# plt.fill_between(smooth_x, smooth_y,alpha=0.2,color='r')
plt.savefig('E:/yuan/pic/loss.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
reward3 = np.load('./data/data_result/reward_{}.npy'.format(3))
reward10 = np.load('./data/data_result/reward_{}.npy'.format(10))
reward9 = np.load('./data/data_result/reward_{}.npy'.format(9))
# Episode = np.mean([len(reward3),len(reward10),len(reward9)])
Episode = 20000
# figsize=(6, 5)
plt.figure(figsize=(8, 6))
mode = 'valid'
step = 3
reward3 = reward3[::4]
reward10 = reward10[::step]
reward9 = reward9[::step]
# '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'
color1, color2, color3 ='#1f77b4', '#ff7f0e', '#2ca02c'
n1,n2,n3 = 50,30,30
plt.plot(np.linspace(0,Episode,len(reward3)),reward3,alpha=0.3,c=color1,linewidth=1)
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward3,n1, mode))),np_move_avg(reward3,n1, mode),c=color1,label='北京')
plt.plot(np.linspace(0,Episode,len(reward10)),reward10,alpha=0.3,c=color2,linewidth=1)
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward10,n2, mode))),np_move_avg(reward10,n2, mode),c=color2,label='广州')
plt.plot(np.linspace(0,Episode,len(reward9)),reward9,alpha=0.3,c=color3,linewidth=1)
plt.plot(np.linspace(0,Episode,len(np_move_avg(reward9,n3, mode))),np_move_avg(reward9,n3, mode),c=color3,label='纽约')
fontsize = 16
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.legend(fontsize=fontsize)
plt.xticks(fontsize=fontsize,fontfamily='Times New Roman')
plt.yticks(fontsize=fontsize,fontfamily='Times New Roman')
plt.xlabel('迭代次数',fontsize=fontsize,fontweight='normal')
plt.ylabel('累积奖励',fontsize=fontsize,fontweight='normal')
# plt.fill_between(smooth_x, smooth_y,alpha=0.2,color='r')
# plt.savefig('F:/yuan/journal/1.5Coalitional Double Auction for Ride-Sharing with Buyout Ask Price/figures/reward.png',dpi=300,bbox_inches='tight')
# bbox_inches='tight'
plt.savefig('E:/yuan/pic/reward.png',dpi=300,bbox_inches='tight')
plt.show()
```
